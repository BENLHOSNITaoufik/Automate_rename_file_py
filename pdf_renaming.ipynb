```python
from pdfminer.high_level import extract_text
import re
from bidi.algorithm import get_display
import os
```

```python
def extract_text_from_first_page(pdf_path):

    # Extract first words from files
    text = extract_text(pdf_path, page_numbers=[1])

    # modify the words
    translate = get_display(text).strip() +'.pdf'
    
    return translate

# specify our folder 
dir = r"\Documents\reading_time\very_short_intro\\"

# convert folder content to the list
files = os.listdir(dir)

for path in  files :

    # join the name of the file and the absolute path
    origin = os.path.join(dir, path)

    # process the path with the function
    text = extract_text_from_first_page(r"\reading_time\very_short_intro\\" + path  )   
    
    new = os.path.join(dir, text)
   
    try:
        os.rename(origin, new)
        print(f"File renamed from '{path}' to '{text}'")

    except FileNotFoundError as e:
        print(f"FileNotFoundError: {e}")
    
```
